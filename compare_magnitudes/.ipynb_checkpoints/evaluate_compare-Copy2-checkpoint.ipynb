```python
import nengo
import nengo_spa as spa
import numpy as np
from itertools import product
import pytry
import matplotlib.pyplot as plt
import pickle
from os import path
from scipy.stats import sem

```

```python


class ExperimentRun(pytry.NengoTrial):
    def params(self):
        self.param('scaling of the number of neurons', n_neurons_scale=1)
        self.param('integrator evidence strength', s_evidence=20)
        self.param('vector dimensions', D=None)
        self.param('whether to reset the comparison integrator at the beginning of each trial', integrator_reset=True)
        self.param('simulation time', T=.2)
        self.param('stimulus to compare', stimulus='')
        self.param('number of samples', n_samples=1000)

    def make_n_neurons_per_dim(self, p):
        # Number of neurons
        n_neurons_per_dim = { # initialize to default values then scale
            module: int(number*p.n_neurons_scale) for module,number in { 
            'AM': 50,
            'Ensemble': 50,
            'State': 50,
            'BG': 100,
            'thal_1': 50,
            'thal_2': 40,
            'Scalar': 50 # this is for the whole module (one dim)
        }.items()}

        return n_neurons_per_dim

    def model(self, p):
        
        rng = np.random.RandomState(p.seed)
        
        vocab = spa.Vocabulary(p.D)
        stimuli = ['D'+str(s) for s in range(2,9)]
        rng.shuffle(stimuli)
        vocab.populate(';'.join(stimuli))
        print('stimuli', stimuli)
        
        n_neurons_per_dim = self.make_n_neurons_per_dim(p)
        with spa.Network(seed=p.seed) as model:
            inp = spa.Transcode(
                lambda t: 'D'+str(p.stimulus),
                output_vocab=vocab
            )
            COM = DirectCompareProcessor(
                vocab, 
                n_neurons_per_dim, 
                rng,
                p.n_samples
            )
            inp >> COM.input
            self.probes = {'compared': nengo.Probe(COM.compared, synapse=.01)}
        return model

    def evaluate_behaviour(self, sim, p, seed):

        print('compare shape', sim.data[self.probes['compared']].shape)
        mean_comparison = np.mean(sim.data[self.probes['compared']])
        return dict(mean_comparison=mean_comparison)

    def evaluate(self, p, sim, plt):
        sim.run(p.T)
        data = self.evaluate_behaviour(sim, p, p.seed)
        return data
    
class DirectCompareProcessor(spa.Network):

    def __init__(
        self, 
        vocab, 
#         s_evidence,
#         feedback,
#         feedback_synapse,
#         reset_period,
        n_neurons_per_dim,
        rng,
        n_samples,
#         tau=.05,
#         integration_threshold=0,
        input_AM_thr=0,
#         decision_thr=.75, 
#         sender=True,
#         receiver=True,
        add_ON=True,
        **kwargs
        ):

        super(DirectCompareProcessor, self).__init__(**kwargs)

        with self:
            
            n_neurons = int(n_neurons_per_dim['Ensemble'])

            D = vocab.dimensions

            self.a = spa.Transcode(input_vocab=vocab, output_vocab=vocab, label="a")
            self.clean_b = nengo.Node([1]) # constant input for our task (5)
            self.input = self.a.input # for our task

            self.clean_a = self.a
            
            train_vectors = np.zeros((n_samples, 2*D))
            train_coords = np.zeros((n_samples, 1))
            xs, ys = rng.choice(range(2,8+1), size=n_samples), rng.choice(range(2,8+1), size=n_samples)
            for i in range(n_samples):
                train_vectors[i, :D] = vocab.parse("D"+str(xs[i])).v
                train_vectors[i, D:] = vocab.parse("D"+str(ys[i])).v
                train_coords[i, 0] = -1 if xs[i]<ys[i] else 1 if xs[i]>ys[i] else 0

            self.combined = nengo.Ensemble(2*D*n_neurons, 2*D)
            nengo.Connection(self.clean_a.output, self.combined[:D])
            nengo.Connection(self.clean_b, self.combined[D:], function=lambda x: x*vocab.parse('D5').v)

            self.compared = nengo.Node(size_in=1)
            nengo.Connection(
                self.combined,
                self.compared,
                function=train_coords,
                eval_points=train_vectors,
                scale_eval_points=False,
                synapse=None
              )

#             self.control_signal = spa.networks.selection.Thresholding(
#                 n_neurons=n_neurons,
#                 n_ensembles=2,
#                 threshold=integration_threshold,
#                 function=lambda x: float(x>0)+.02
#             )
#             nengo.Connection(self.compared, self.control_signal.input[0])
#             nengo.Connection(self.compared, self.control_signal.input[1], transform=-1) # if input is negative


#             # Create integrator
#             self.integrator = nengo.Ensemble(n_neurons*40, 2, radius=1.5)
#             nengo.Connection( # input to the integrator
#                 self.compared,
#                 self.integrator[0],
#                 transform=tau*s_evidence,
#                 synapse=tau
#               )
#             nengo.Connection( # input to the control signal
#                 self.control_signal.output[0],
#                 self.integrator[1],
#                 synapse=tau
#               )

#             nengo.Connection( # input to the control signal (if input is negative)
#                 self.control_signal.output[1],
#                 self.integrator[1],
#                 synapse=tau
#               )

#             nengo.Connection(self.integrator, self.integrator[0], # feedback
#                 synapse=tau,
#                 function=lambda x: x[0]*x[1] # control
#             )      

#             if reset_period is not None:
#                 self.integrator_reset = nengo.Node(lambda t: t%reset_period<.2) # reset integrator at the beginning of each trial, during 200 ms (should be at least tau seconds)
#                 nengo.Connection(
#                     self.integrator_reset,
#                     self.integrator.neurons,
#                     synapse=None,
#                     transform=-100*np.ones((self.integrator.n_neurons, 1)),
#                 )

#             add_ON = "+ON" if add_ON else ""
#             self.output = spa.ThresholdingAssocMem(
#                 decision_thr, 
#                 output_vocab, 
#                 mapping={k:k+add_ON for k in ["LESS", "MORE"]},
#                 function=lambda x: x>0,
#                 n_neurons = n_neurons_per_dim['AM']

#             )

#             nengo.Connection(self.integrator[0], self.output.input, 
#                 function=lambda x: -x*output_vocab.parse("LESS").v+x*output_vocab.parse("MORE").v
#             )

#             self.preconscious = self.output.output


        self.declare_input(self.input, vocab) # for our task
#         self.declare_output(self.output.output, vocab)
```

```python
data_dir = 'evaluate_compare/all'
```

```python
seeds = list(range(30))
n_neurons_scales = [.5,1]#,1.5]#,2]
Ds = [96,192]
stimuli = [2,4,6,8]
ns_samples = [100,1000]#,10000]

mean_comparisons = np.zeros((len(seeds), len(n_neurons_scales), len(Ds), len(stimuli), len(ns_samples)))

i=0
for condition in product(*tuple([range(len(param_values)) for param_values in [seeds, n_neurons_scales, Ds, stimuli, ns_samples]])):
    i+=1
    print('condition ', i, '/', mean_comparisons.size, ':', condition)    
    
    data_filename = 'run_seed'+str(seeds[condition[0]])+'_nscale'+str(n_neurons_scales[condition[1]])+'_D'+str(Ds[condition[2]])+'_stim'+str(stimuli[condition[3]])+'_nsamples'+str(ns_samples[condition[4]])
    
    try:
        mean_comparison = pickle.load(open(data_dir+'/'+data_filename, "rb"))
    
    except FileNotFoundError:
        try:
            mean_comparison = ExperimentRun().run(
                seed=seeds[condition[0]],
                n_neurons_scale=n_neurons_scales[condition[1]],
                D=Ds[condition[2]],
                stimulus=stimuli[condition[3]],
                n_samples=ns_samples[condition[4]],
                backend="nengo_ocl",
                data_dir=data_dir,
                data_filename=data_filename,
            )['mean_comparison']

            pickle.dump(mean_comparison, open(data_dir+'/'+data_filename, "wb"))
        
        except MemoryError:
            print('Memory error', seeds[condition[0]], n_neurons_scales[condition[1]], Ds[condition[2]], stimuli[condition[3]], ns_samples[condition[4]])
            mean_comparison = np.nan
    
    mean_comparisons[condition] = mean_comparison
    print(mean_comparison)



```

```python
print(np.min(mean_comparisons))
```

```python
x_spacing = .1
x_spacing_count = (-x_spacing * (len(n_neurons_scales) * len(Ds) * len(ns_samples)-1))/2
linestyles = ['-','-.','--']
colors = ['black','gray','black']
for n_neurons_scale in range(len(n_neurons_scales)):
    for D in range(len(Ds)):
        for n_samples in range(len(ns_samples)):
            
            plt.errorbar(
                x_spacing_count+np.array(stimuli),
                np.abs(np.mean(mean_comparisons[:,n_neurons_scale,D,:,n_samples], axis=(0,))),
                np.std(mean_comparisons[:,n_neurons_scale,D,:,n_samples], axis=(0,)),
                label=str(n_neurons_scales[n_neurons_scale])+' '+str(Ds[D])+' '+str(ns_samples[n_samples]),
                linestyle=linestyles[n_neurons_scale],
                color=colors[n_samples]
            )
            
            x_spacing_count += x_spacing
            
plt.xticks([2,4,6,8])
plt.xlabel('Stimulus')
plt.ylabel('Comparison magnitude')
    
plt.legend()
plt.show()
```
