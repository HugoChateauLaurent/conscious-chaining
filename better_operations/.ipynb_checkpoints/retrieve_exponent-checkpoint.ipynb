```python
import nengo_spa as spa
import nengo

use_ocl = False
if use_ocl:
    import nengo_ocl
    sim_cls = nengo_ocl.Simulator
else:
    sim_cls = nengo.Simulator
    
import numpy as np
import matplotlib.pyplot as plt


```

```python
dimensions = 64*2
vocab = spa.Vocabulary(dimensions)
vocab.populate("BASE.unitary()")

base = vocab.parse("BASE")
train_limit = 5
test_limit = train_limit + 5
n_samples = 100
neurons_per_dim = 50
T = 10

def e_in(t):
    if t<T/2:
        return np.sin(4*t)*train_limit
    else:
        return np.sin(4*t)*test_limit

def power(e):
    return np.fft.ifft(np.fft.fft(base.v) ** e).real


train_vectors = np.zeros((n_samples, dimensions))
train_coords = np.zeros((n_samples, 1))


for i in range(n_samples):
    x = np.random.uniform(low=-train_limit, high=train_limit)
    train_vectors[i, :] = power(x)
    train_coords[i, 0] = x
    
    
with spa.Network() as model:
    
    e = nengo.Node(e_in, label="e")
    
    y = nengo.Node(lambda t,x: power(x), size_in=1, label="y")
    nengo.Connection(e, y)
    
    y_ens = nengo.Ensemble(dimensions * neurons_per_dim, dimensions, radius=train_vectors.max(), label="y_ens")
    nengo.Connection(y, y_ens)
    
    retrieved_e = nengo.Ensemble(neurons_per_dim, 1, radius=test_limit * 1.4, label="retrieved_e")
    nengo.Connection(
        y_ens,
        retrieved_e,
        function=train_coords,
        eval_points=train_vectors,
        scale_eval_points=False,
      )

    probes = {obj.label: nengo.Probe(obj, synapse=.005) for obj in [e, y, y_ens, retrieved_e]}


with sim_cls(model) as sim:
    sim.run(T)
```

```python

plt.figure(figsize=(16,6))

plt.ylim(-10,10)
plt.xlim(0,T)
plt.xlabel("t")
plt.title("Retrieved from ensemble")
plt.plot(sim.trange(), sim.data[probes["retrieved_e"]], label="retrieved")
plt.plot(sim.trange(), sim.data[probes["e"]], color="black", linestyle="--", label="e")
plt.legend()




plt.show()

```
