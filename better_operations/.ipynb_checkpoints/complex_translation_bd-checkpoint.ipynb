```python
import nengo_spa as spa
import nengo
import numpy as np
import matplotlib.pyplot as plt

dimensions = 4
vocab = spa.Vocabulary(dimensions)
vocab.populate("BASE.unitary()")

base = vocab.parse("BASE").v # the base vector
scale = 1.00000000001

def power(base, e): # as in: https://forum.nengo.ai/t/vector-exponentials-with-unitary-semantic-pointers/505/2
    y = np.fft.ifft(np.fft.fft(scale*base) ** e)
    return y

def retrieve_e(y, base):
    return np.log(np.absolute(np.fft.fft(y))) / np.log(np.absolute(np.fft.fft(scale*base)))


for e in np.linspace(-10,10):
    y = power(base, e)
    print(np.linalg.norm(base), np.linalg.norm(y))
    retrieved_e = retrieve_e(y, base) # works with e=0,1 maybe also e=2, but not the rest. And I would expect more consistency along dimensions

    print(np.round(e,2), np.round(retrieved_e,2))
```
