```python
import nengo
import nengo_spa as spa
import numpy as np
import matplotlib.pyplot as plt
```

```python
Ds = np.linspace(1,500,dtype=int)
seeds = 1000
dots_A_B = np.zeros((seeds,Ds.size))
dots_A_C = np.zeros((seeds,Ds.size))
dots_diff = np.zeros((seeds,Ds.size))

for i in range(seeds):
    for j,D in enumerate(Ds):
        vocab = spa.Vocabulary(D)
        vocab.populate('A.unitary() ; B=A*A ; C=B*A')
        dots_A_B[i,j] = np.abs(np.dot(vocab.parse('A').v, vocab.parse('B').v))
        dots_A_C[i,j] = np.abs(np.dot(vocab.parse('A').v, vocab.parse('C').v))
        dots_diff[i,j] = dots_A_B[i,j] - dots_A_C[i,j]
```

```python
plt.figure(figsize=(18,4))

plt.subplot(1,3,1)
plt.errorbar(Ds, np.mean(dots_A_B, axis=0), dots_A_B.std(axis=0))
plt.xlabel("D")
plt.title("| dot(A, A^2) |")

plt.subplot(1,3,2)
plt.errorbar(Ds, np.mean(dots_A_C, axis=0), dots_A_C.std(axis=0))
plt.xlabel("D")
plt.title("| dot(A, A^3) |")

plt.subplot(1,3,3)
plt.errorbar(Ds, np.mean(dots_diff, axis=0), dots_diff.std(axis=0))
plt.xlabel("D")
plt.title("dot(A, A^2) - dot(A, A^3)")


plt.show()
```
