```python
import nengo
import nengo_spa as spa
import pytry
import numpy as np
from random import shuffle
import random
import sys, os
import math
from IPython import display
from scipy.optimize import brute, minimize
from scipy.stats import sem
from scipy.stats import linregress
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.style as style

import pickle

import os
from itertools import product

# color-blind accessibility
from cycler import cycler
colors = ['#006BA4', '#FF800E', '#ABABAB', '#595959', '#5F9ED1', '#C85200', '#898989', '#A2C8EC', '#FFBC79', '#CFCFCF']
default_cycler = cycler('color', colors)
plt.rc('axes', prop_cycle=(default_cycler))

# Import our classes
from ADD_model import ExperimentRun, create_vocab
```

# Optimization and model parameters

```python
TAU_PREV = .001
TAU_AMPROC = .001
ALPHA = .2
GW_SCALE = 20 # scale the input of GW for saturation and catching very short visual stimuli
BG_THR = .1
BG_BIAS = .5


data_dir = "many_steps"
```

```python
ns_ADD_processors = np.array(list(range(2,22,2)) + [50])
seeds = np.array(range(50))
n_neurons_scales = [1]
Ds = [int(16*(i+1)) for i in range(1,8,2)]
print(ns_ADD_processors)
print(Ds)

accs = np.zeros((len(seeds), len(ns_ADD_processors), len(n_neurons_scales), len(Ds)))
conditions = product(*tuple([range(len(param_values)) for param_values in [seeds, ns_ADD_processors, n_neurons_scales, Ds]]))

for i,condition in enumerate(conditions):
    print('condition ', i+1, '/', accs.size, ':', condition)  
    print('\t Simulating...')
    data_filename = 'run_seed'+str(seeds[condition[0]])+'_nproc'+str(ns_ADD_processors[condition[1]])+'_scale'+str(n_neurons_scales[condition[2]])+'_D'+str(Ds[condition[3]])

    try:
        with open(data_dir+'/'+data_filename+'.pickle', 'rb') as handle:
            acc = pickle.load(handle)

    except FileNotFoundError:
        results = ExperimentRun().run(
            n_neurons_scale=n_neurons_scales[condition[2]],
            vocab=create_vocab(Ds[condition[3]], seeds[condition[0]], ns_ADD_processors[condition[1]]),
            tau_AMproc=TAU_AMPROC,
            tau_prev=TAU_PREV,
            alpha=ALPHA,
            GW_scale=GW_SCALE,
            n_ADD_proc=ns_ADD_processors[condition[1]],
            BG_thr=BG_THR,
            BG_bias=BG_BIAS,
            seed=seeds[condition[0]],
            backend="nengo_ocl",
            data_dir=data_dir,
            data_filename=data_filename,
            plt=False
        )
        acc = results['acc']
        with open(data_dir+'/'+data_filename+'.pickle', 'wb') as handle:
            pickle.dump(acc, handle, protocol=pickle.HIGHEST_PROTOCOL)

    accs[condition] = acc



```

```python

ratio = 6
f,(ax,ax2) = plt.subplots(1,2,sharey=False, facecolor='w', gridspec_kw={'width_ratios': [ratio, 1]})

# plot the same data on both axes
for i,D in reversed(list(enumerate(Ds))):
    ax.plot(ns_ADD_processors, accs[:,:,:,i].mean(axis=0) * 100, label=D, linewidth=4, alpha=.7)
    ax2.plot(ns_ADD_processors, accs[:,:,:,i].mean(axis=0) * 100, label=D, linewidth=4, alpha=.7)
    
    # write final accuracy
    ax2.text(ns_ADD_processors[-1]+.2, -1+accs[:,:,:,i].mean(axis=0)[-1] * 100, str(int(accs[:,:,:,i].mean(axis=0)[-1,0] * 100))+'%')

ax.set_ylabel('Accuracy (%)', size=12)
f.text(0.5, 0.02, 'Number of additions', ha='center', size=12)
ax.legend(title=r'$\mathcal{D}$')

ax.set_xlim(1,21)
ax2.set_xlim(49,51)
ax.set_ylim(-2,102)
ax2.set_ylim(-2,102)

# hide the spines
ax.spines['right'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax2.spines['left'].set_visible(False)
ax2.spines['top'].set_visible(False)

ax.set_xticks(ns_ADD_processors[:-1])
ax2.set_xticks([ns_ADD_processors[-1]])
ax2.set_yticks([])

d = .015 # how big to make the diagonal lines in axes coordinates
# arguments to pass plot, just so we don't keep repeating them
kwargs = dict(transform=ax.transAxes, color='k', clip_on=False)
ax.plot((1-d,1+d), (-d,+d), **kwargs)

kwargs.update(transform=ax2.transAxes)  # switch to the bottom axes
ax2.plot((-d*ratio,+d*ratio), (-d,+d), **kwargs)
plt.savefig('figures/many_accuracy.pdf', format='pdf')

plt.show()
```

```python
ps = np.zeros((len(Ds), len(ns_ADD_processors)))
for i,ref in enumerate(ns_ADD_processors):
    ref_acc = accs[:,:,:,:].mean(axis=0)[i]
    ps[:,i] = np.power(ref_acc, 1/ref)

ps[np.where(ps==0)] = np.nan
# ps[np.where(ps==1)] = np.nan
ps = np.nanmean(ps, axis=1)

observed = accs[:,:,0,:].mean(axis=0) * 100
predicted = np.asarray([np.power(ps, n) for n in ns_ADD_processors]) * 100

slope, intercept, r, p, stderr = linregress(predicted.flatten(), observed.flatten())
print(slope, intercept, r, p, stderr)
# plt.plot([0,100], np.array([0,100]), color='black', linewidth=.5, linestyle='-')
plt.plot([0,100], np.array([0,100])*slope+intercept, color='black', linewidth=1.5, linestyle='dotted')

# print([np.power(ps, ns_ADD_processors[i]) for n in ns_ADD_processors])
for i,D in reversed(list(enumerate(Ds))):
    plt.scatter(predicted[:,i], observed[:,i], s=40, label=D, alpha=.7)
    
plt.xlabel('Predicted accuracy (%)', size=12)
plt.ylabel('Observed accuracy (%)', size=12)
plt.xlim(-2,102)
plt.ylim(-2,102)
ax = plt.gca()
ax.set_aspect('equal', adjustable='box')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.legend(title='$\mathcal{D}$')
plt.savefig('figures/many_prediction.pdf', format='pdf')
plt.show()
```
