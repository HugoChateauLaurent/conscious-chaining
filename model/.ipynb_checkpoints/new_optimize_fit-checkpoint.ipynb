# Optimization and model parameters

```python
D = int(96)  # the dimensionality of the vectors
ALPHA = .2
GW_SCALE = 20 # scale the input of GW for saturation and catching very short visual stimuli
BG_THR = .1
BG_BIAS = .5
STIM_DURATION = .029
FIX_DURATION = .25
T_ANSWER = 1
N_NEURONS_SCALE = 1
INTEGRATOR_RESET = True
N_SAMPLES = 10000

TAU_PREV = .001
TAU_AMproc = .001

STARTING_SEED = 1
N_SEEDS = 10

N_BLOCKS_PER_OPERATION = 1 # default: 10
N_TRIALS_PER_DIGIT = 2 # default: 5
N_DIFFERENT_DIGITS = 4 # default: 4
N_DIFFERENT_OPERATIONS = 3 # default: 3

data_dir = "evaluate_fit/new_optimization4"
```

```python
# opt_res = skopt.load(data_dir+'/optimization_results.pickle')

def create_evaluation_function(
    loss='rmse', 
    plot=False, 
    save_plot=False, 
    starting_seed=STARTING_SEED, 
    n_seeds=N_SEEDS, 
    n_trials_per_digit=None, 
    compare_errorates=True, 
    compare_RTs=True, 
    dim_idx=None, 
    save_csv=False, 
    override_existing=False):
    
    if n_trials_per_digit is None:
        n_trials_per_digit = N_TRIALS_PER_DIGIT
    
    def delayed_eval(delay, data):
            
        delayed_data = data.df.copy(deep=True)
        delayed_data['rt'] += delay[0] * 1000
        delayed_data = Data(delayed_data)

        error = delayed_data.mean_differences_error(compare_errorrates=compare_errorates, compare_RTs=compare_RTs, tasks=range(N_DIFFERENT_OPERATIONS))
#         print('\t', delay, error)

        return error
    
    def evaluation_function(params):
    
        print("params:", params)
        data = []
        
        if dim_idx is None:
            dim = D
        else:
            dim = int(params[dim_idx] * 16)
    
#         print('Simulating...')
        for i,seed in enumerate(range(starting_seed, starting_seed+n_seeds)):
#             print('\t seed '+str(i+1)+'/'+str(n_seeds))
            parameter_filename = '_crosstalk'+str(params[0])+'_tauprev'+str(params[1])+'_taucompare'+str(params[2])+'_combscale'+str(params[3])+'_D'+str(dim)    
            data_filename = 'run_seed'+str(seed)+parameter_filename
            
            try:
                assert not override_existing
                seed_data = pd.read_pickle(data_dir+'/'+data_filename+'.csv')

            except (FileNotFoundError, AssertionError) as e:
                xp = create_xp(1, N_BLOCKS_PER_OPERATION, n_trials_per_digit, N_DIFFERENT_DIGITS, N_DIFFERENT_OPERATIONS, STIM_DURATION, FIX_DURATION, T_ANSWER, seed)
                results = ExperimentRun().run(
                    s_crosstalk=params[0],
                    tau_prev=params[1],
                    tau_AMproc=TAU_AMproc,
                    tau_compare=params[2],
                    n_neurons_scale_combined=params[3],
                    n_neurons_scale=N_NEURONS_SCALE,
                    n_samples=N_SAMPLES,
                    vocab=create_vocab(dim, seed),
                    xp=xp,
                    integrator_reset=INTEGRATOR_RESET,
                    alpha=ALPHA,
                    GW_scale=GW_SCALE,
                    BG_thr=BG_THR,
                    BG_bias=BG_BIAS,
                    seed=seed,
                    backend="nengo_ocl",
                    data_dir=data_dir,
                    data_filename=data_filename,
                    plt=False
                )

                seed_data = results['data'].df

                seed_data.to_pickle(data_dir+'/'+data_filename+'.csv')
            data.append(seed_data)


        data = Data(pd.concat(data))
        
        if loss == 'rmse':
            delay_filename = 'delay'+parameter_filename
            try:
                assert not override_existing
                best_delay = np.load(data_dir+'/'+delay_filename+'.npy')

            except (FileNotFoundError, AssertionError) as e:
                best_delay = minimize(delayed_eval, (.25,), bounds=((0,None),), args=data)['x']
            
            print('best delay:', best_delay)

            error = delayed_eval(best_delay, data)
            data.df['rt'] += best_delay*1000
            np.save(data_dir+'/'+delay_filename, best_delay)
            
        elif loss == 'error rate':
            error = data.error_rate * 100
        
        print(loss, ':', error)
        print("error rate:", str(round(data.error_rate*100,1))+'%')

        if plot:
            save_file = data_dir+'/'+data_filename if save_plot else None
            data.plot_fig2_simple(plot_humans=True, save_file=save_file, errorates=compare_errorates)
            data.plot_fig2_chained(plot_humans=True, save_file=save_file, errorates=compare_errorates)
            
        if save_csv:
            data.df.to_csv(data_dir+'/'+'dataframe'+parameter_filename+'.csv', sep='\t', index=False)
        
        return error
    
    return evaluation_function
```

# Optimise response times

```python
RTs_res = gp_minimize(create_evaluation_function(compare_errorates=False, plot=False), 
    dimensions=[
        Real(0, 1, name=r"$\omega_{V \rightarrow COM}$"), # crosstalk
        Real(.001, .01, name=r"$\tau_r$ (ms)"), # tau prev (in seconds but changed to ms in plot)
        Real(0.005, .2, name=r"$\tau_c$ (ms)"), # tau compare (in seconds but changed to ms in plot)
        Real(.1, 1, name=r"$\mathcal{C}$"), # scaling of the number of neurons in the combined state
    ],
    x0=[0.2, .001, 0.02, 0.25],
    n_calls=200,
    random_state=STARTING_SEED,
    n_random_starts=20,
               
)
```

```python
print(RTs_res.keys())
for key in ['x','fun', 'func_vals', 'x_iters']:
    np.save(key+'.npy', RTs_res[key])
```

```python
def optimisation_metrics(opt_res):
    regret_axes = plot_regret(opt_res)
    regret_axes.set_title('')
    regret_axes.get_lines()[0].set_color("black")
    regret_axes.set_ylabel(r"$\sum_{i=0}^n(\mathcal{L}(\theta, \theta_s) - optimum)$ after $n$ calls", size=12)
    regret_axes.set_xlabel(regret_axes.get_xlabel(), size=12)
    plt.grid(False)
    regret_axes.spines['right'].set_visible(False)
    regret_axes.spines['top'].set_visible(False)
    plt.savefig('figures/regret.eps', format='eps', bbox_inches = 'tight')
    plt.show()

    convergence_axes = plot_convergence(opt_res)
    convergence_axes.set_title('')
    convergence_axes.get_lines()[0].set_color("black")
    convergence_axes.set_ylabel(r"$\min \mathcal{L}(\theta, \theta_s)$ after $n$ calls", size=12)
    convergence_axes.set_xlabel(convergence_axes.get_xlabel(), size=12)
    plt.grid(False)
    convergence_axes.spines['right'].set_visible(False)
    convergence_axes.spines['top'].set_visible(False)
    plt.savefig('figures/convergence.eps', format='eps', bbox_inches = 'tight')
    plt.show()

optimisation_metrics(RTs_res)
```

```python

# plot_evaluations(RTs_res)
plot_objective_axes = plot_objective(RTs_res, n_samples=40)

tau_r_moreticks = np.asarray([.002,.004,.006,.008,.01])
tau_r_fewticks = np.asarray([.002,.004,.006,.008])
plot_objective_axes[1,0].set_yticklabels((tau_r_fewticks*1000).astype(int))
plot_objective_axes[1,1].set_xticklabels((tau_r_moreticks*1000).astype(int))
plot_objective_axes[3,1].set_xticklabels((tau_r_fewticks*1000).astype(int))


tau_c_moreticks = np.asarray([0,.04,.08,.12,.16,.2])
tau_c_fewticks = np.asarray([.04,.08,.12,.16])
plot_objective_axes[2,0].set_yticklabels((tau_c_fewticks*1000).astype(int))
plot_objective_axes[2,2].set_xticklabels((tau_c_moreticks*1000).astype(int))
plot_objective_axes[3,2].set_xticklabels((tau_c_fewticks*1000).astype(int))

for i in range(4):
    plot_objective_axes[i,i].get_lines()[0].set_color("black")
     
plt.savefig('figures/plot_objective.eps', format='eps', bbox_inches = 'tight')
plt.show()
```
