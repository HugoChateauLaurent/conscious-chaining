# Optimization and model parameters

```python
D = int(96)  # the dimensionality of the vectors
PROC_FDBCK = .9
PROC_FDBCK_SYN = .005
GW_FDBCK = 1
GW_SCALE = 20 # scale the input of GW for saturation and catching very short visual stimuli
BG_THR = .1
BG_BIAS = .5


data_dir = "evaluate_performance"
```

```python
ns_ADD_processors = np.array(range(2,20,2))
print(ns_ADD_processors)
seeds = np.array(range(10))
n_neurons_scales = [.5,1,2]

accs = np.zeros((len(ns_ADD_processors), len(seeds), len(n_neurons_scales)))

for i,condition in enumerate(product(*tuple([range(len(param_values)) for param_values in [ns_ADD_processors, seeds, n_neurons_scales]]))):
    print('condition ', i+1, '/', accs.size, ':', condition)  
    print('\t Simulating...')
    data_filename = 'run_seed'+str(seeds[condition[1]])+'_nproc'+str(ns_ADD_processors[condition[0]])+'_scale'+str(n_neurons_scales[condition[2]]) 

    try:
        with open(data_dir+'/'+data_filename+'.pickle', 'rb') as handle:
            acc = pickle.load(handle)

    except FileNotFoundError:
        results = ExperimentRun().run(
            n_neurons_scale=n_neurons_scales[condition[2]],
            vocab=create_vocab(D, seeds[condition[1]], ns_ADD_processors[condition[0]]),
            proc_feedback=PROC_FDBCK,
            proc_feedback_synapse=PROC_FDBCK_SYN,
            GW_feedback=GW_FDBCK,
            GW_scale=GW_SCALE,
            n_ADD_proc=ns_ADD_processors[condition[0]],
            BG_thr=BG_THR,
            BG_bias=BG_BIAS,
            seed=seeds[condition[1]],
            backend="nengo_ocl",
            data_dir=data_dir,
            data_filename=data_filename,
            plt=False
        )
        acc = results['acc']
        with open(data_dir+'/'+data_filename+'.pickle', 'wb') as handle:
            pickle.dump(acc, handle, protocol=pickle.HIGHEST_PROTOCOL)

    accs[condition] = acc



```

```python
for i,scale in enumerate(n_neurons_scales):
    plt.plot(ns_ADD_processors, accs[:,:,i].mean(axis=1), label=scale)
    plt.show()
```
